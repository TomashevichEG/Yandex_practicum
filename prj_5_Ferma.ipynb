–ü–∞—Ä–∞–º–µ—Ç—Ä—ã '–í–∫—É—Å_–º–æ–ª–æ–∫–∞','–ñ–∏—Ä–Ω–æ—Å—Ç—å_%','–ë–µ–ª–æ–∫_%' –Ω–µ —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏–∑—É—é—Ç —Ü–µ–ª–µ–≤–æ–π –ø—Ä–∏–∑–Ω–∞–∫ –∑–∞ –£–¥–æ–π_–∫–≥.

```python
#–£–±–µ—Ä–µ–º –Ω–µ—Ü–µ–ª–µ–≤—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏ –¥–ª—è –ø–µ—Ä–≤–æ–π –º–æ–¥–µ–ª–∏.
data1=data.drop(['id','–í–∫—É—Å_–º–æ–ª–æ–∫–∞','–ñ–∏—Ä–Ω–æ—Å—Ç—å_%','–ë–µ–ª–æ–∫_%','–ò–º—è_–ø–∞–ø—ã'],axis=1)
```
